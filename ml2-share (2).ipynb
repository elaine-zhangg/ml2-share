```python
# filling unrated content rating by using euclidean distance of the plot keywords doc2vec values
def infer_content_rating(df):
    """
    Output the assigned file filling all the unrated entries by finding the content rating inferring from
    plot keywords using euclidean distance.
    """   
    remove_missing = df.copy()
    # Find rows with missing content ratings
    missing_rows = df[(df['content_rating'] == 'Not Rated') | (df['content_rating'] == 'Unrated')]
    remove_missing = remove_missing.drop(missing_rows.index)
    plot_keywords = np.array(remove_missing['plot_keywords'])
    float_value = 0.323324
    distances = np.abs(plot_keywords - float_value)
    return distances

infer_content_rating(train_x_processed)
```

```python
# filling unrated content rating by comparing euclidean distance of the plot keywords doc2vec values
def infer_content_rating(df):
    """
    Output the assigned file filling all the unrated entries by finding the content rating inferring from
    plot keywords using euclidean distance.
    """   
    remove_missing = df.copy()
    missing_rows = df[(df['content_rating'] == 'Not Rated') | (df['content_rating'] == 'Unrated')]
    remove_missing = remove_missing.drop(missing_rows.index)
    plot_keywords = np.array(remove_missing['plot_keywords'])
    for index, row in missing_rows.iterrows():
        distances = np.abs(plot_keywords - row['plot_keywords'])
        closest_index = distances.argmin()
        inferred_rating = df.iloc[closest_index]['content_rating']
        df.at[index, 'content_rating'] = inferred_rating  
    return df

infer_content_rating(train_x_processed)
infer_content_rating(test_processed)
train_x['content_rating'] = train_x_processed['content_rating']
test['content_rating'] = test_processed['content_rating']
```

```python
# conducting one hot encoding to language, country, actor3 names and content rating for both training and test 
# data for models that cannot process non- numerical data

def one_hot_encoding(parameter, file):
    """
    return the input file converting the assigned parameter to an array by one hot encoding
    """
    encoder = OneHotEncoder()
    array = np.array(file[parameter]).reshape(-1, 1)
    encoded = encoder.fit_transform(array)
    onehot_array = encoded.toarray()
    file[parameter]= onehot_array.tolist()
    return file

names = ['language', 'actor_3_name', 'country', 'content_rating']
for i in names:
    one_hot_encoding(i, train_x_processed)
    one_hot_encoding(i, test_processed)
    
train_x_processed
```
